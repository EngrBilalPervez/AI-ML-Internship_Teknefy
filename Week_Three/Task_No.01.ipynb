```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df=pd.read_csv("task1_dataset.csv")

# Cleaning the data,applying feature scaling 
print(pd.isna(df).sum())
x=df[["YearsExperience"]]
y=df[["Salary"]]

# Visualize the data on a scatter plot (Experience on X-axis, Salary on Y-axis).
plt.scatter(x,y)
plt.title("This is visualizing the given data through scatter plot")
plt.xlabel("Experience x_axis")
plt.ylabel("Salary y_axis")
```

```python
# Split the data into Training and Testing sets (80% Train, 20% Test).

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(f"This is x_axis training data below\n{x_train}\nAND\n This is x_axis Testing data below\n{x_test}")
print(f"This is y_axis training data below\n{y_train}\nAND\n This is y_axis Testing data below\n{y_test}")
```

```python
# Train the LinearRegression model.
model=LinearRegression()
model.fit(x,y)
```

```python
# Prediction: Ask the model: If a person has 5 years of experience, what should their salary be?
num=float(input("Enter a experience do you have."))
a=[[num]]
predicted_salary=model.predict(a)
print(f"You enter experience is {num} so the prediction salary of this experience is {predicted_salary}")
```

```python
# Draw the "Best Fit Line" over the scatter plot
plt.scatter(x, y, label="Actual Data")
plt.plot(x, model.predict(x), label="Best Fit Line",color="red",lw=4)
plt.xlabel("Years of Experience x_axis")
plt.ylabel("Salary y_axis")
plt.title("Best Fit Line")
plt.legend()
```
