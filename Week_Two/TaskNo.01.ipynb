```python
import numpy as np
# Array creation 
array_1d=np.array([1,2,3,4,5]) # 1D array creation
print(f"This is 1D array = {array_1d}")

arr_2d=np.array([[1,2,3,4],
                 [6,7,8,9],
                 [10,11,12,14]]) # 2D array creation
print(f"This is 2D array =\n{arr_2d}")

arr_zero=np.zeros((3,3))  # 3x3 zeros array creation
print(f"This is all elements of matrix 3x3 is zero \n{arr_zero}")

arr_one=np.ones((2,3))  # 2x3 ones array creation
print(f"This is all elements of matrix 2x3 is ones \n{arr_one}")
same=np.full((3,4),7) # 3x4 same number array creation
print(f"All elements of this matrix is same number\n {same}")
rge=np.arange(0,21,2) #Only create 1D array  even number 0 to 20
print(f"0 to 20 Even number is \n {rge}")
```

```python
# Array Indexing & Slicing
array_1d=np.array([1,2,3,4,5]) #Indexing 1D array 
print(f"Get the value of index 3 = {array_1d[3]}")

arr_2d=np.array([[1,2,3,4],
                 [6,7,8,9],
                 [10,11,12,14]])
get=arr_2d[1,2] #Get the value of row 2 col 3 value 
print(f"This is row2 col3 value = {get}")
get_col=arr_2d[:, 1] #Get the only value of col 2
print(f"value of col 2 is {get_col}")
```

```python
# Vectorized Mathematical Operations
array_1=np.array([1,2,3,4,5])
array_2=np.array([6,7,8,9,0])
total= array_1+array_2
print(f"Sum of 2 1D vector is = \n{total}") 
mul= array_1 * array_2
print(f"Multiplication of 2 1D vector is = \n{mul}")

# Aggregation (Summary Statistics)
#Find maximum number of two vectors
m=np.concatenate((array_1,array_2))
maximun=np.max(m)
print(f"Multiplication of 2 1D vector is = \n{maximun}")
#Find manimum number of two vectors
minimum=np.min(m)
print(f"Multiplication of 2 1D vector is = \n{minimum}")
```

```python
# Reshaping Arrays
arr=np.arange(1,17)
print(f"1D array is = {arr}")
#Reshape the 1D array in 4x4
reshaped=arr.reshape(4,4)
print(f"4x4 array is = \n{reshaped}")
```

```python
# Matrix Operations
array_1=np.array([1,2,3,4,5])
array_2=np.array([6,7,8,9,0])
# Dot Product
dot_p=np.dot(array_1,array_2)
print(f"Dor product of two vector is {dot_p}")

#Transpose of 2D  
arr_2d=np.array([[1,2,3,4],
                 [6,7,8,9],
                 [10,11,12,14]])
tra=arr_2d.T
print(f"Transpose of 2D vector is \n{tra}")           
```

```python
# Broadcasting
arr_2d=np.array([[1,2,3],
                 [6,7,5]])
a=np.array([1,2,3])
result= arr_2d + a
print(f"Adding \n{result}")
```
