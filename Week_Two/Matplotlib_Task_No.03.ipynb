```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

arr1=np.array([1,2,5,4,2,6,4,3,9])
arr2=np.array([10,11,12,13,14,15,16,17,18])

# Creating Basic plot
plt.plot(arr1,arr2,color="red",linewidth=8) #Also apply line styles

#Titles and Labels
plt.title("This is my first Matplotlib Chart") 
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
# Now creating a Scatter plot
a1=np.array([1,2,3,4,5])
a2=np.array([4,3,2,8,5])
plt.scatter(a1,a2,color="green",linewidth=10)  #Also apply line chart style

#Titles and Labels
plt.title("This is my first Scatter Chart") 
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
# Now creating histogram chart
f=np.array([1,0,3.1,4.5])
plt.hist(f,color="pink",linewidth=0.9)
plt.show()
```

```python
# Creating Pie char
s = [3, 30, 20, 10]
l = ["A", "B", "C", "D"]

plt.pie(s, labels=l)
plt.title("This is Pie Chart")
plt.figure(figsize=(0.8,0.5))  #Applying also figure sizes
plt.show()
```

```python
# Crreating legend
plt.plot([1,2,3], [2,1,6], label="Student A",color="red",linewidth=7) #Also applying line styles
plt.plot([1,2,3], [1,3,2], label="Student B",color="blue",linewidth=3)
plt.title("This is my firt legend plot") #Also applying title 
plt.legend()
plt.show()
```

```python
# Now creating the Subplots
plt.subplot(2,2,1)
plt.plot(arr1,arr2)
plt.subplot(2,2,4)
plt.plot(a1,a2)
plt.subplot(2,2,2)
plt.plot([1,2,3,4],[2,3,1,2])
plt.subplot(2,2,3)
plt.plot([1,10,18,3,2],[3,5,2,8,10])

# Also apply the saving plot chart
plt.savefig("Saving_Plot.png")

```

```python

#Griding 
plt.plot(a1, a2)
plt.grid(True)
plt.show()
```
