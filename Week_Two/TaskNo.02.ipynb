```python
import numpy as np
import pandas as pd

# Data Loading & Saving (I/O Operations)
df=pd.read_csv('students_data.csv')
print(df)
#Adding new column,Adress
df['Address']="Pakistan"
print("\nAfter Adding the new column\n",df)
#Exporting Clean Data
df.to_csv('clean_data.csv')
```

```python
# Data Inspection & Understanding
print(f"Following operation are performing here \nHead = {df.head()},\nShape = {df.shape}")
```

```python
# Data Selection (Selecting Rows & Columns)

print(df['Name']) #Single column selection
print("\nMultiple columns selecting.\n\n",df[['Name','Age','Grade']])
print(f"Row selecting \n{df.loc[3]}")
```

```python
# Indexing & Slicing
# selecting one row only
print(f"\n{df.loc[1]}")
# Selecting multiple rows
print(f"\n\nMultiples Rows indexing.\n{df.loc[[1,4]]}")

#Indexing base selection
print(f"This is indexing base selection.\n{df.iloc[1:3]}")
```

```python
# Data Cleaning & Preprocessing
#Remove a column address
print(f"Removing Address Column\n{df.drop('Address',axis=1)}")
df['Address']="Pakistan"
print(f"\n\nAdding Address Column\n{df}")
```

```python
# Sorting Operations
print(f"Sorting value only one column\n{df.sort_values('Age')}")
```

```python
# Vectorized Operations (Column-Wise Math)
df["Age"] = df["Age"] + 10
df["Student_ID"]=df["Student_ID"]*2
print(df)
```

```python
# Apply, Map, Transform (Function Application)

# Employee data dictionary
employee_data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Ahmed", "Bilal", "Sara", "Ayesha", "Hamza"],
    "Age": [25, 30, 28, 26, 32],
    "Department": ["HR", "IT", "Finance", "Marketing", "IT"],
    "Salary": [50000, 60000, 55000, 52000, 65000]
}
n=pd.DataFrame(employee_data)
print(n)
n["Category(Young/Old)"] = df["Age"].apply(lambda x: "Young" if x < 27 else "Old")
n["Gender"] = df["Name"].map({"Bilal": "Male", "Ahmed": "Male","Hamza":"Male","Ayesha":"Female","Sara":"Female"})
print(f"\n\nAfter Applying the operation \n{n}")
```

```python
# Missing Data Handling
data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Ahmed", "Bilal", np.nan, "Ayesha", "Hamza"],
    "Age": [25, 19, 28, 26, 32],
    "Department": ["HR", "IT", "Finance", np.nan, "IT"],
    "Salary": [50000, 60000, np.nan, 52000, 65000]
}
d=pd.DataFrame(data)
print(d)

#Now performing the different operation
print(f"\nMissing Value\n{d.isna()}")
print(f"\nRemoving Value\n{d.dropna(thresh=4)}")
print(f"\nAdding Value\n{d.fillna("Please_Fill")}")
```
