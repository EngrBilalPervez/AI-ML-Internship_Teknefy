```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Now fetch the by default data in seaborn 
df=sns.load_dataset("tips")

# Creating a Scatter Plot
sns.scatterplot(x='tip',y='size',data=df)
plt.title("This is my first seaborn scatter plot")
```

```python
# Now creating the Line plot
sns.lineplot(x='total_bill',y='tip',data=df)
```

```python
# Now create a bar plot
sns.barplot(x='tip',y='size',data=df,hue=True,palette="rainbow")
```

```python
#Now creating a count plot
sns.countplot(x='sex',data=df)
```

```python
#Now ceate a Histogram (displot / histplot)
sns.histplot(df['total_bill'],bins=40,kde=True)
```

```python
#Now create a box plot
sns.boxplot(x='tip',y='total_bill',data=df)
```

```python
#Now create a violin plot
sns.violinplot(x='total_bill',y='tip',data=df)
```

```python
#Pairplot
sns.pairplot(df)
```

```python
#Now create a jointplot
sns.jointplot(df)
```

```python
#Now create a dSistribution plots
sns.histplot(df['total_bill'], kde=True,bins=45)
```
